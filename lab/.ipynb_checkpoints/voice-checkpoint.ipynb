```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import numpy as np

# Initialize the pipeline
pipeline = KPipeline(lang_code='a')
text = "Testing Kokoro TTS in JupyterLab on macOS!"

# Get audio from the generator
audio_generator = pipeline(text, voice='af_bella')
audio_chunks = list(audio_generator)  # Collect all chunks into a list

# Combine chunks into a single NumPy array
audio = np.concatenate(audio_chunks)  # Assumes chunks are NumPy arrays

# Reshape for mono audio (if needed)
if audio.ndim == 1:
    audio = audio.reshape(-1, 1)

# Save and play
sf.write("output.wav", audio, 24000)
display(Audio("output.wav"))
```

```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import numpy as np

# Initialize the pipeline
pipeline = KPipeline(lang_code='a')
text = "Testing Kokoro TTS in JupyterLab on macOS!"
audio_generator = pipeline(text, voice='af_bella')
audio_chunks = list(audio_generator)

# Get the Result object (single chunk)
result = audio_chunks[0]

# Extract audio (use result[2] or result.audio)
audio = result[2]  # The tensor from the tuple
# Alternatively: audio = result.audio  # Uncomment to test if preferred

# Convert PyTorch tensor to NumPy and reshape for mono
audio = audio.cpu().numpy()  # Move to CPU and convert
if audio.ndim == 1:
    audio = audio.reshape(-1, 1)  # Ensure 2D for soundfile

# Save and play
sf.write("output.wav", audio, 24000)
display(Audio("output.wav"))
```

```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import numpy as np

pipeline = KPipeline(lang_code='a')
text = "Testing Kokoro TTS in JewpeterLab on macOS!"
audio_generator = pipeline(text, voice='af_bella')
result = list(audio_generator)[0]
audio = result[2].cpu().numpy().reshape(-1, 1)

# Test different sample rates
for rate in [16000, 22050, 24000, 44100]:
    filename = f"output_{rate}.wav"
    sf.write(filename, audio, rate)
    print(f"Saved {filename} at {rate} Hz")
    display(Audio(filename))
    
```

```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import numpy as np

pipeline = KPipeline(lang_code='a')
text = "Testing Kokoro TTS in JewpterLab on macOS!"
audio_generator = pipeline(text, voice='af_bella')
result = list(audio_generator)[0]
audio = result[2].cpu().numpy()

# Normalize to [-1, 1]
audio = audio / np.max(np.abs(audio)) if np.max(np.abs(audio)) > 0 else audio
audio = audio.reshape(-1, 1)

sf.write("output_normalized.wav", audio, 22050)
display(Audio("output_normalized.wav"))
```
